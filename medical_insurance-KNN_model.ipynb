```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

#  Import and read the insurance.csv.
import pandas as pd 
insurance_df = pd.read_csv('Resources/insurance.csv')
insurance_df.head()
```

```python
# Cleaning DataFrame - drop extra index column
insurance_df = insurance_df.drop(columns = ['index'])
insurance_df
```

```python
# Cleaning DataFrame - cleaning charges column
insurance_df['charges'] = insurance_df['charges'].map("{:,.2f}".format)
insurance_df
```

```python
insurance_df.dtypes
```

```python
# Define features set
X = insurance_df.copy()
X.drop("charges", axis=1, inplace=True)
X.head()
```

```python
# Define target vector
y = insurance_df["charges"].values.reshape(-1, 1)
y[:5]
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
```

```python
X[:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=1)
X_train.shape
```

```python
#NN
```

```python
# Import our dependencies
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import sklearn as skl
import tensorflow as tf
```

```python
# Create scaler instance
X_scaler = skl.preprocessing.StandardScaler()

# Fit the scaler
X_scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create the Keras Sequential model
nn_model = tf.keras.models.Sequential()
```

```python
# Add our first Dense layer, including the input layer
nn_model.add(tf.keras.layers.Dense(units=5, activation="relu", input_dim=2))
```

```python
# Add the output layer that uses a probability activation function
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
```

```python
# Check the structure of the Sequential model
nn_model.summary()
```

```python
# Compile the Sequential model together and customize metrics
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])

# Fit the model to the training data
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=100)
```

```python
# Create a DataFrame containing training history
history_df = pd.DataFrame(fit_model.history)

# Increase the index by 1 to match the number of epochs
history_df.index += 1

# Plot the loss
history_df.plot(y="loss")
plt.show()
```

```python
# Plot the accuracy
history_df.plot(y="accuracy")
plt.show()
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
