```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.preprocessing import StandardScaler

#  Import and read the insurance.csv.
import pandas as pd 
insurance_df = pd.read_csv('Resources/insurance.csv')
insurance_df.head()
```

```python
# Cleaning DataFrame - drop extra index column
insurance_df = insurance_df.drop(columns = ['index'])
insurance_df
```

```python
# Cleaning DataFrame - cleaning charges column
insurance_df['charges'] = insurance_df['charges'].map("{:,.2f}".format)
insurance_df
```

```python
insurance_df.dtypes
```

```python
# Define features set
X = insurance_df.copy()
X.drop("charges", axis=1, inplace=True)
X.head()
```

```python
# Define target vector
y = insurance_df["charges"].values.reshape(-1, 1)
y[:5]
```

```python
# Encode the categorical variables using get_dummies
X = pd.get_dummies(X)
```

```python
X[:]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    random_state=27)
X_train.shape
```

```python
#confusion matrix
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(solver='lbfgs', random_state=10)
classifier
```

```python
# Train the data
classifier.fit(X_train, y_train)
```

```python
# Score the model
print(f"Training Data Score: {classifier.score(X_train, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test, y_test)}")
```

```python
# Predict outcomes for test data set
predictions = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predictions)
```
